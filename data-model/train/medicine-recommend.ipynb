```python
#chia mỗi chuỗi trong cột 'Description' của DataFrame medicines thành các danh sách các từ.
medicines['Description'].apply(lambda x:x.split())
```

```python
#chia reason và des thành các từ
medicines['Reason'] = medicines['Reason'].apply(lambda x:x.split())
medicines['Description'] = medicines['Description'].apply(lambda x:x.split())
```

```python
#loại bỏ khoảng trắng trong từng từ của mỗi danh sách trong cột 'Description'
medicines['Description'] = medicines['Description'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#gộp des + reason = tags
medicines['tags'] = medicines['Description'] + medicines['Reason'] 
```

```python
new_df = medicines[['index','Drug_Name','tags']]
```

```python
#chuyển đổi mỗi danh sách các từ trong cột 'tags' thành một chuỗi
new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english',max_features=5000)
```

```python
#Tách chuỗi đầu vào thành các từ riêng biệt,
def stem(text):
  y = []
# mặc định là tách theo khoảng trắng
  for i in text.split():
    y.append(ps.stem(i))
#Các từ gốc được thu thập vào danh sách y
  return " ".join(y)  
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# sử dụng CountVectorizer để biến đổi các nhãn từ văn bản trong cột 'tags' của new_df thành một ma trận đếm các từ
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
#truy xuất các từ vựng đã học được từ dữ liệu sử dụng CountVectorizer trong biến cv.
feature_names = cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#so sánh sự tương đồng giữa các văn bản hoặc các đặc trưng văn bản.
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(medicine):
    #tìm chỉ số của thuốc
    medicine_index = new_df[new_df['Drug_Name'] == medicine].index[0]
    #Tính toán độ tương đồng
    distances = similarity[medicine_index]
    #Sắp xếp và đề xuất các thuốc tương tự
    medicines_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])
    
    for i in medicines_list:
        print(new_df.iloc[i[0]].Drug_Name)
    
```

```python
recommend("Paracetamol 125mg Syrup 60mlParacetamol 500mg Tablet 10'S")
```
